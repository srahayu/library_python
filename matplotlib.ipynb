```python
import numpy as np # untuk menghitung matematika 
import matplotlib.pyplot as plt #grafik 
```

# cara menggunakan matplotlib 
1. membuat data 
2. membuat plot 
3. menampilkan plot 

```python
#1.membuat data (sin(2wt + theta)
#camel case (punuk unta) 

def sinusGenerator (amplitudo,frekuensi,tAkhir,theta):
    t = np.arange(0,tAkhir,0.1)
    y = amplitudo * np.sin(2*frekuensi*t + np.deg2rad(theta))
    return t,y
```

```python
#membuat plot 
#grafik 1
t1,y1 = sinusGenerator(1,1,4,0)
plt.plot(t1,y1)
#grafik 2
t2,y2 = sinusGenerator(1,1,4,30)
plt.plot(t2,y2,'y')

t3,y3 = sinusGenerator (1,1,4,60)
plt.plot(t3,y3,'b--')

t4,y4 = sinusGenerator (1,1,4,90)
plt.plot(t4,y4,'r-o')


#tanda -o di dekat huruf r menghaslkan grafik dengan ada lingkaran pada grafiknya 
#tanda -- di dekat huruf b menghasilkan garis putus-putus pada grafik 
#menampilkan plot 
plt.show()
```

```python
def sinusGenerator (amplitudo,frekuensi,tAkhir,theta):
    t = np.arange(0,tAkhir,0.1)
    y = amplitudo * np.sin(2*frekuensi*t + np.deg2rad(theta))
    return t,y
t1,y1 = sinusGenerator(1,1,4,0)
t2,y2 = sinusGenerator(1,1,4,90)

plt.plot (t1,y1, label="sin(0)")
plt.plot (t2,y2, label="sin(90)")
plt.legend()
plt.show
```

# membuat teks didalam plot 

```python
def sinusGenerator (amplitudo,frekuensi,waktu,theta):
    t = np.arange(0,waktu,0.1)
    y = amplitudo * np.sin(2*frekuensi*t + np.deg2rad(theta))
    return t,y
t,y = sinusGenerator(1,1,4,0)
plt.plot(t,y)

plt.text(2,0.5, r'$ y = \mathcal{A}.sin(2 \omega t)$')
plt.text(2,0.4, r'$ \mathcal{A} = 1 cm, \omega = 1 Hz$')

plt.show()
```

# cara mengatur ticks nya sesuai dengan titik-titik yang benar/ penting

```python
sudut = np.arange (0,360,1)
y = np.sin (np.deg2rad(sudut))

plt.plot(sudut,y)
plt.ylabel("magnituda")
plt.xlabel ("sudut")

plt.yticks ([-1,0,1])
plt.xticks (
    [0,90,180,270,360],
    [r'${0}^0$',r'${90}^0$',r'${180}^0$',r'${270}^0$',r'${360}^0$'])
plt.show()
```

# cara menggeser garis absis ke tengah 

```python
sudut = np.arange (0,360,1)
y = np.sin (np.deg2rad(sudut))

plt.plot(sudut,y)
plt.title ("grafik sinusoidal")
plt.text(190,1,"magnituda")
plt.text(360,0.1,"sudut")
plt.yticks ([-1,-0.5,0,0.5,1])
plt.xticks (
    [0,90,180,270,360],
    [r'${0}^0$',r'${90}^0$',r'${180}^0$',r'${270}^0$',r'${360}^0$'])

ax = plt.gca()
ax.spines ['left'].set_position(('data',180))
ax.spines ['bottom'].set_position(('data',0))
ax.spines ['right'].set_color('none')
ax.spines ['top'].set_color('none')
plt.show()
```
